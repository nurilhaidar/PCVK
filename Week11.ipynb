```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

**<h3>1. TEMPLATE MATCHING<h3>**

```python
img = cv.imread('/content/drive/MyDrive/Images/Modul11/cats_and_bunnies.jpg')
template = cv.imread('/content/drive/MyDrive/Images/Modul11/cat2_templatejpg.jpg', cv.IMREAD_GRAYSCALE)

w,h = template.shape[::-1]
```

```python
img_CCOEFF = img.copy()
img_CCOEFF_GRAY = cv.cvtColor(img_CCOEFF, cv.COLOR_BGR2GRAY)
res = cv.matchTemplate(img_CCOEFF_GRAY, template, cv.TM_CCOEFF)
minmin_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)

top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)

cv.rectangle(img_CCOEFF,top_left, bottom_right, [0,0,255], 5)

fig, ax = plt.subplots(1,2)
ax[0].imshow(res, cmap='gray')
ax[0].set_title('Matching Result')
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].imshow(cv.cvtColor(img_CCOEFF, cv.COLOR_BGR2RGB))
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_title('Detected Point')
```
