```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

1. GAMMA CORRECTION

```python
print('Gamma correction pada citra')
print('---------------------------')
img = cv.imread('/content/drive/MyDrive/Images/female.tiff')
try:
  gamma = int(input('Masukkan nilai gamma : '))
except ValueError:
    print('Nilai yang anda masukkan salah !')

def gammaCorrection(image, value):
  value /= 10
  image = cv.cvtColor(image, cv.COLOR_BGR2RGB).astype(np.float32)/255
  gamma_img = np.power(image, value)
  return gamma_img

img_gamma = gammaCorrection(img, gamma)

plt.subplot(1,2,1)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1,2,2)
plt.imshow(img_gamma)
plt.title(f'Image With {gamma/10} Gamma')
```

3. PSNR atau Average Denoising

```python
import glob
from math import log10, sqrt
from google.colab.patches import cv2_imshow

# Jumlah Citra Average 5

original = cv.imread('/content/drive/MyDrive/Images/galaxy.jpg')
noise = []
for img in glob.glob('/content/drive/MyDrive/Images/noises/*.jpg'):
    n = cv.imread(img)
    noise.append(n)

img_noise = noise[0]
noise_number = 4
for i in range(noise_number):
    if i == 0:
        pass
    else:
        alpha = 1.0 / (i + 1)
        beta = 1.0 - alpha
        img_noise = cv.addWeighted(noise[i], alpha, img_noise, beta, 0.0)

def psnr(img1, img2):
    mse = np.mean((img1-img2) ** 2)
    if(mse == 0):
      return 100

    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse))
    return psnr

psnr = psnr(original, img_noise)
print(f'Nilai PSNR adalah : {psnr}')
frame = cv.hconcat((original, img_noise))
cv2_imshow(frame)
```

4. Masking

```python
# gambar original
ktp = cv.imread('/content/drive/MyDrive/Images/KTP.jpeg')
ktp = cv.cvtColor(ktp, cv.COLOR_BGR2RGB)

# membuat blank frame
blank = np.zeros(ktp.shape, ktp.dtype)

# membuat frame mass
mask = cv.circle(blank, center=(610, 220), radius=30, color=(255,255,255), thickness=cv.FILLED)
mask = cv.circle(blank, center=(730, 220), radius=30, color=(255,255,255), thickness=cv.FILLED)

# Mask menggunakan operator AND
mask_img = cv.bitwise_and(ktp, mask)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(ktp)

plt.subplot(1,3,2)
plt.imshow(mask)

plt.subplot(1,3,3)
plt.imshow(mask_img)
```

```python
# Mask menggunakan operator OR
mask_or = cv.bitwise_not(mask)

mask_img = cv.bitwise_or(ktp, mask_or)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(ktp)

plt.subplot(1,3,2)
plt.imshow(mask)

plt.subplot(1,3,3)
plt.imshow(mask_img)
```

```python
# Mask menggunakan operator NAND
mask_img = cv.bitwise_not(cv.bitwise_and(ktp, mask))

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(ktp)

plt.subplot(1,3,2)
plt.imshow(mask)

plt.subplot(1,3,3)
plt.imshow(mask_img)
```

```python
# Mask menggunakan operator XOR
mask_img = cv.bitwise_xor(ktp, mask)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(ktp)

plt.subplot(1,3,2)
plt.imshow(mask)

plt.subplot(1,3,3)
plt.imshow(mask_img)
```

```python
# Mask menggunakan operator XOR
mask_img = cv.bitwise_not(ktp, mask)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(ktp)

plt.subplot(1,3,2)
plt.imshow(mask)

plt.subplot(1,3,3)
plt.imshow(mask_img)
```

Kesimpulan :

- NOT : Menggabungkan kedua gambar, kemudian membalikkan warna dari penggabunngan sebeumnya
-  OR : Menggabungkan kedua gambar, namun masking image tidak ditampilkan
- AND : Menggabungkan kedua gambar, namun outputnya berupa gambar pada lingkaran masking
- NAND : hasil sama seperti AND namun warna dibalikkan (inverse)
- XOR : menggabungkan kedua gambar, namun warna pada masking di balikkan(inverse)
